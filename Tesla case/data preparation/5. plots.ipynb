```python
import pandas as pd
import numpy as np
from datetime import datetime

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
### import the data
tesla_data = pd.read_csv('4.tesla_twitter_data_aggregated_prices.csv')
#tesla_data = tesla_data.drop(columns=['Subjuctivity','Polarity'])
tesla_data.head()
```

```python
tesla_data['date'] = pd.to_datetime(tesla_data['date'])
tesla_data.head()
```

```python
tesla_data.columns
```

```python
tesla_data['total_count_tweet_scaled'] =  (tesla_data['total_count_tweet'] - tesla_data['total_count_tweet'].min()) / (tesla_data['total_count_tweet'].max() - tesla_data['total_count_tweet'].min())
```

```python
tesla_data['Close_scaled'] =  (tesla_data['Close'] - tesla_data['Close'].min()) / (tesla_data['Close'].max() - tesla_data['Close'].min())
```

```python
tesla_data['Polarity_scaled'] =  (tesla_data['Polarity'] - tesla_data['Polarity'].min()) / (tesla_data['Polarity'].max() - tesla_data['Polarity'].min())
```

```python
tesla_data['Class_sentiment_negative_scaled'] =  (tesla_data['Class_sentiment_negative'] - tesla_data['Class_sentiment_negative'].min()) / (tesla_data['Class_sentiment_negative'].max() - tesla_data['Class_sentiment_negative'].min())
```

```python
tesla_data['Class_sentiment_positive_scaled'] =  (tesla_data['Class_sentiment_positive'] - tesla_data['Class_sentiment_positive'].min()) / (tesla_data['Class_sentiment_positive'].max() - tesla_data['Class_sentiment_positive'].min())
```

```python
tesla_data.plot(x="date", y=["total_count_tweet_scaled", "Close_scaled"], figsize=(10, 5), grid=True)
plt.title('plot of total counts of tweets vs close price per date, Tesla')
plt.show()
```

```python
tesla_data.plot(x="date", y=["Polarity_scaled", "Close_scaled"], figsize=(10, 5), grid=True)
plt.title('plot of polarity vs close price per date, Tesla')
plt.show()
```

```python
tesla_data.plot(x="date", y=["Class_sentiment_positive_scaled", "Close_scaled"], figsize=(10, 5), grid=True)
plt.title('plot of Class_sentiment_positive vs close price per date, Tesla')
plt.show()
```

```python
tesla_data.plot(x="date", y=["Class_sentiment_negative_scaled", "Close_scaled"], figsize=(10, 5), grid=True)
plt.title('plot of Class_sentiment_negative vs close price per date, Tesla')
plt.show()
```
