# get preprocessed data

```python
data_sentim_ana = pd.read_csv('2.amazon_data_dates_sentiment_analysis.csv')
data_sentim_ana
```

# calculate the sum of feature like, comment, retweet

```python
tl0 = data_sentim_ana['like_num'][data_sentim_ana['date']==data_sentim_ana.date[93]].sum() # date 01-01-2015 starts
tl1 = data_sentim_ana['like_num'][data_sentim_ana['date']==data_sentim_ana.date[94]].sum() # date 02-01-2015 starts
tl0,tl1
```

```python
data_smedia_sum = data_sentim_ana.groupby(data_sentim_ana.date).sum()
data_smedia_sum.head()
```

#### calculate the total number of tweetts per day then sum it with the sum of re tweet for that day

```python
twt_ct_per_day = data_sentim_ana.date.groupby(data_sentim_ana.date).count()
twt_ct_per_day.values
```

```python
data_smedia_sum['tweet_count'] = twt_ct_per_day.values
data_smedia_sum
```

```python
total = []
count = 0
for i in data_smedia_sum.index:
        total.append(data_smedia_sum['tweet_count'][i] + data_smedia_sum['retweet_num'][i])
        print(data_smedia_sum['tweet_count'][i] + data_smedia_sum['retweet_num'][i])
```

```python
data_smedia_sum['total_count_tweet'] = total
```
