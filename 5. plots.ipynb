```python
import pandas as pd
import numpy as np
from datetime import datetime

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
### import the data
amazon_data = pd.read_csv('4.amazon_twitter_data_aggregated_prices.csv')
#tesla_data = tesla_data.drop(columns=['Subjuctivity','Polarity'])
amazon_data.head()
```

```python
amazon_data['date'] = pd.to_datetime(amazon_data['date'])
amazon_data.head()
```

```python
amazon_data.columns
```

```python
amazon_data['total_count_tweet_scaled'] =  (amazon_data['total_count_tweet'] - amazon_data['total_count_tweet'].min()) / (amazon_data['total_count_tweet'].max() - amazon_data['total_count_tweet'].min())
```

```python
amazon_data['Close_scaled'] =  (amazon_data['Close'] - amazon_data['Close'].min()) / (amazon_data['Close'].max() - amazon_data['Close'].min())
```

```python
amazon_data['Polarity_scaled'] =  (amazon_data['Polarity'] - amazon_data['Polarity'].min()) / (amazon_data['Polarity'].max() - amazon_data['Polarity'].min())
```

```python
amazon_data['Class_sentiment_negative_scaled'] =  (amazon_data['Class_sentiment_negative'] - amazon_data['Class_sentiment_negative'].min()) / (amazon_data['Class_sentiment_negative'].max() - amazon_data['Class_sentiment_negative'].min())
```

```python
amazon_data['Class_sentiment_positive_scaled'] =  (amazon_data['Class_sentiment_positive'] - amazon_data['Class_sentiment_positive'].min()) / (amazon_data['Class_sentiment_positive'].max() - amazon_data['Class_sentiment_positive'].min())
```

```python
amazon_data.plot(x="date", y=["total_count_tweet_scaled", "Close_scaled"], figsize=(10, 5), grid=True)
plt.title('plot of total counts of tweets vs close price per date, Amazon')
plt.show()
```

```python
amazon_data.plot(x="date", y=["Polarity_scaled", "Close_scaled"], figsize=(10, 5), grid=True)
plt.title('plot of polarity vs close price per date, Amazon')
plt.show()
```

```python
amazon_data.plot(x="date", y=["Class_sentiment_positive_scaled", "Close_scaled"], figsize=(10, 5), grid=True)
plt.title('plot of Class_sentiment_positive vs close price per date, Amazon')
plt.show()
```

```python
amazon_data.plot(x="date", y=["Class_sentiment_negative_scaled", "Close_scaled"], figsize=(10, 5), grid=True)
plt.title('plot of Class_sentiment_negative vs close price per date, Amazon')
plt.show()
```
